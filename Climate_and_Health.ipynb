```python
# Load Datasets
import pandas as pd
import numpy as np

try:
    df_health = pd.read_csv("Climate_and_Health.csv")
    print("Climate and Health Data loaded successfully. Shape:", df_health.shape)
except FileNotFoundError:
    print("Error: Climate_and_Health.csv not found.")

try:
    df_solar = pd.read_csv("solar_energy_climate_health_dataset.csv")
    print("Solar Energy Data loaded successfully. Shape:", df_solar.shape)
except FileNotFoundError:
    print("Error: solar_energy_climate_health_dataset.csv not found.")
```

```python
# Clean and Preprocess Data

# 1. Health Data Cleaning
if 'df_health' in locals():
    # Convert date to datetime
    if 'date' in df_health.columns:
        df_health['date'] = pd.to_datetime(df_health['date'])
        print("Converted 'date' to datetime type in Health Data.")

    # Drop duplicates
    initial_rows = len(df_health)
    df_health = df_health.drop_duplicates()
    print(f"Dropped {initial_rows - len(df_health)} duplicates from Health Data.")

    # Handle missing values (Checking only)
    missing_count = df_health.isnull().sum().sum()
    print(f"Total missing values in Health Data: {missing_count}")
    # df_health = df_health.dropna() # Uncomment if you want to drop missing values immediately

    print("Health Data Head:")
    display(df_health.head())

# 2. Solar Data Cleaning
if 'df_solar' in locals():
    initial_rows_solar = len(df_solar)
    df_solar = df_solar.drop_duplicates()
    print(f"Dropped {initial_rows_solar - len(df_solar)} duplicates from Solar Data.")
    
    print("Solar Data Head:")
    display(df_solar.head())
```
